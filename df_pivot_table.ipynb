```python
import pandas as pd
import numpy as np
```

```python
data = {
    'fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Banana','Orange'],
    'current_price': [1500, 2980, 5500, 3600,3000,4520],
    'month': ['February', 'June', 'March','March','February','June'],
    'estimated_future_price':[2500,3100,8200,6210,3685,7400]
}

df = pd.DataFrame(data)
df
```

```python
df.pivot_table(index=['fruit','month'])
```

```python
data = {
    'fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Banana'],
    'price': [1.0, 0.5, 0.8, 1.2, 0.6],
    'expiry_date': ['2024-04-10', '2024-04-09', '2024-04-12', '2024-04-11', '2024-04-10'],
    'arrival_date': ['2024-04-01', '2024-04-02', '2024-04-03', '2024-04-01', '2024-04-02']
}

df = pd.DataFrame(data)

# Converting date strings to datetime objects
df['expiry_date'] = pd.to_datetime(df['expiry_date'])
df['arrival_date'] = pd.to_datetime(df['arrival_date'])

# Displaying the DataFrame
print("Original DataFrame:")
print(df)

# Using pivot_table method
pivot_df = df.pivot_table(index='fruit', values='price')

# Displaying the pivot table
print("\nPivot Table:")
pivot_df
```

```python
df = pd.DataFrame({'Number': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'Letter': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'Number_2': [1, 2, 3, 4, 5, 6],
                   'Letter_2': ['x', 'y', 'z', 'q', 'w', 'e'],
                   'Number_3':[7,8,9,10,11,12]
                   })
df
```

```python
df.pivot(index=['Number'], columns='Letter', values=['Number_2'])
```

```python
df.pivot(index=['Number','Letter_2'], columns='Letter', values=['Number_2'])
```

```python
df.pivot(index=['Number'], columns='Letter', values=['Number_2','Number_3'])
```

```python
data = {
    'date': ['2024-04-01', '2024-04-01', '2024-04-02', '2024-04-02'],
    'category': ['A', 'B', 'A', 'B'],
    'value1': [10, 20, 30, 40],
    'value2': [100, 200, 300, 400]
}

df = pd.DataFrame(data)
df
```

```python
df.pivot_table(index='date',columns='category',values=['value1','value2'])
```

```python
data={
    'Name':['Nongshim','Shin','Nongshim','Shin','Shin','Nongshim'],
    'Type':['shrimp','potato','shrimp','potato','potato','shrimp'],
    'Price':[2500,1700,2300,1300,600,1200]
}
df=pd.DataFrame(data)
df
```

```python
df.pivot_table(index=['Name','Type'])
```

```python
df.pivot_table(index=['Name'], columns='Type')
```

```python
df.pivot_table(index=['Name','Type'], margins=True, margins_name='Umumiy', aggfunc='sum')
```

```python
df.pivot_table(index=['Name','Type'], margins=True, margins_name='Umumiy', aggfunc=['sum','mean'])
```

```python
df.pivot_table(index=['Name','Type'], aggfunc='sum')
```
