```python
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Value': [10, 20, 30, 40]
}
df = pd.DataFrame(data)
df
```

```python
pivot_table = pd.pivot_table(df, values='Value', index='Date', columns='Category')
pivot_table
```

```python
pivot_table_with_margins = pd.pivot_table(df, values='Value', index='Date', columns='Category', aggfunc='sum')
pivot_table_with_margins
```

```python
pivot_table_with_margins.columns
```

```python
# Creating a pivot table with handling missing values
pivot_table_fillna = pd.pivot_table(df, values='Value', index='Date', columns='Category', aggfunc='sum', fill_value=0)
pivot_table_fillna
```

```python
import pandas as pd
import numpy as np
url='https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/master/world_cities_air_water_quality.csv'
df=pd.read_csv(url)
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
columns=['City','Region','Country','AirQuality','WaterPollution']
df.columns=columns
df.head()
```

```python
# 1-usul
df['Region']=df['Region'].str.replace('"','')
df.head()
# 2-usul
#df['Region']=df['Region'].apply(lambda x : x.replace('"',''))
#df.head()
```

```python
df.loc[4,'Region']
df['Region']=df['Region'].str.strip()
df.loc[4,'Region']
df.loc[318,'Region']
```

```python
df.loc[311:316]
```

```python
df['Country'].value_counts()
df['Country']=df['Country'].str.replace('"','')
df.head()
```

```python
df.loc[257,'Country']
df['Country']=df['Country'].str.strip()
df.loc[257,'Country']
```

```python
df.head()
```

```python
df['Country']=df['Country'].str.replace('United States of America','USA')
df['Country']=df['Country'].str.replace("People's Republic of China",'China')
df['Country']=df['Country'].str.replace('United Kingdom','UK')
df['Country'].value_counts()
```
