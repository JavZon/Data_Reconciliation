```python
import pandas as pd

arrays = [['bar', 'bar', 'foo', 'foo'],
          ['one', 'two', 'one', 'two']]
index = pd.MultiIndex.from_arrays(arrays, names=['A', 'B'])
df = pd.DataFrame({'values': range(4)}, index=index)
df
```

```python
#df.swaplevel('A', 'B')
#df.swaplevel(-2, -1)
df.swaplevel(0,1)
```

```python
regions = ['North', 'South', 'East', 'West']
months = ['Jan', 'Feb', 'Mar', 'Apr']
index = pd.MultiIndex.from_product([regions, months], names=['Region', 'Month'])
sales_data = pd.DataFrame({'Sales': [100, 120, 110, 105, 90, 95, 130, 125, 115, 105, 100, 110, 120, 115, 105, 100]},
                          index=index)
sales_data
```

```python
sales_data.swaplevel('Region', 'Month')
```
