```python
import numpy as np
import pandas as pd
```

```python
val=[[26,21,20,24],[30,33,21,32],[75,70,66,77],[73,50,40,63]]
col=pd.MultiIndex.from_product([['Air_temperature','Air_humidity'],['January','February']])
ind=pd.MultiIndex.from_arrays([['Uzb','Uzb','USA','USA'],['Tashkent','Samarkand','California','New York']])
reshaped_val = [val[:4], val[4:8], val[8:12], val[12:]]
df=pd.DataFrame(val,index=ind, columns=col)
df
```
