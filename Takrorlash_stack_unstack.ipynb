```python
import pandas as pd
import numpy as np
```

```python
level=[['X','Y','Z'],[1,2,3]]
data=np.random.randint(1,100,9)
ind=pd.MultiIndex.from_product(level, names=['First','Second'])
ser=pd.Series(data, index=ind)
```

```python
ser=pd.Series(data, index=ind)
ser
```

```python
ser['X']
```

```python
ser['Z',2]
```

```python
ser['Y':'Z']
```

```python
ser.loc[['Y','Z'], 2]
```

```python
ser
```

```python
df=ser.unstack()
df
```
