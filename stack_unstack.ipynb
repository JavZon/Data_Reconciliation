```python
index = pd.MultiIndex.from_tuples([('A', 1), ('A', 2), ('B', 1), ('B', 2)], names=['Region', 'Year'])
data = {'Sales': [100, 200, 150, 250]}
df = pd.DataFrame(data, index=index)
df.loc[('A',1)]
```

```python
df.stack().unstack()
```

```python
data = {'A': [1, 2], 'B': [3, 4], 'C': [5, 6]}
df = pd.DataFrame(data)
df
```

```python
df.stack()
```

```python
#So, the unstack() method splits the stacked data (MultiIndex Series) into individual columns,
#essentially turning the stacked data into a wide format DataFrame.
#Remembering this rule can help you recall that unstack() is used to split the stacked data into separate columns.

data=[25,80,36,2]
ind=pd.MultiIndex.from_tuples([('A',1),('A',2),('B',1),('B',2)], names=['First','Second'])
ser=pd.Series(data, ind)
ser
```

```python
df=ser.unstack()
df
```

```python
df.columns
```
