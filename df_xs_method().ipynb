```python
import pandas as pd
import numpy as np
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
index = pd.MultiIndex.from_arrays(arrays, names=('first', 'second'))
df = pd.DataFrame({'data': [1, 2, 3, 4]}, index=index)
```

```python
df
```

```python
df.xs(2, level='second', drop_level=False)
```

```python
df.xs('A', level='first', drop_level=False)
```

```python
import pandas as pd
import numpy as np

d = {'num_legs': [4, 4, 2, 2],
     'num_wings': [0, 0, 2, 2],
     'class': ['mammal', 'mammal', 'mammal', 'bird'],
     'animal': ['cat', 'dog', 'bat', 'penguin'],
     'locomotion': ['walks', 'walks', 'flies', 'walks']}
df = pd.DataFrame(data=d)
df
```

```python
df=df.pivot_table(index=['class','animal','locomotion'])
df.sort_index(inplace=True)
```

```python
df
```

```python
df.xs('mammal',level='class', drop_level=False)
```

```python
df.xs(('mammal','dog','walks'), drop_level=False)
```

```python
df.loc[('mammal','dog','walks')]
```

```python
df.xs(key='mammal')
```

```python
df
```

```python
df.loc[('mammal', slice('cat'), slice('walks'))]
```

```python
df.xs(key=('mammal', 'bat', 'flies'))
```

```python
df
```

```python
df.xs('cat', level=1, drop_level=False)
```

```python
df.xs('bird', level=0, drop_level=False)
```

```python
df.xs('cat', level='animal', drop_level=False)
```

```python
df.xs(('bird', 'walks'),
      level=[0, 2], drop_level=False)
```

```python
df
```

```python
df.xs('num_wings', axis=1)
```

```python
df.xs('num_wings', axis='columns')
```
