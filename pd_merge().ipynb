```python
import pandas as pd
import numpy as np
```

```python
data1 = {'ID': [1, 2, 3, 4],
         'Name': ['Alice', 'Bob', 'Charlie', 'David'],
         'Age': [25, 30, 35, 40]}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {'ID': [1, 3, 5, 7],
         'Product': ['Apple', 'Banana', 'Orange', 'Grapes'],
         'Price': [1.2, 0.5, 2.0, 1.8]}
df2 = pd.DataFrame(data2)
print(df1)
print(df2)
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='inner')
```

```python
df1.merge(df2, left_on='ID', right_on='ID')
```

```python
d1={
    'Name':['foo','bar','zoo','quu'],
    'Price':[25,36,20,14],
    'Quality':[30.2, 12.8, 25.4, 51.2 ]
}
df1=pd.DataFrame(d1)
df1
```

```python
d2={
    'Name_ID':['foo','laa','zoo','muu'],
    'Narx':[25,39,20,99],
    'Sifat':[30.2, 18.3, 25.4, 55.1 ]
}
df2=pd.DataFrame(d2)
print(df1)
print(df2)
```

```python
df1.merge(df2, left_on='Name', right_on='Name_ID')
```

```python
a1={
    'Type':['Bus','Ship','Plane'],
    'Price':[35000,48000,94000]
}
df1=pd.DataFrame(a1)
df1
```

```python
a2={
    'Type':['Train','Plane','Ship'],
    'Fuel_km':[55,3600,240]
}
df2=pd.DataFrame(a2)
print(df1)
print()
print(df2)
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
a2={
    'Type_b':['Train','Plane','Ship'],
    'Fuel_km':[55,3600,240]
}
df2=pd.DataFrame(a2)
df2
```

```python
df1.merge(df2, left_on='Type', right_on='Type_b')
```

```python
df1.merge(df2, left_on='Type', right_on='Type_b', how='outer')
```

```python
a=df1.merge(df2, left_on='Type', right_on='Type_b')
a.drop(columns='Type_b', inplace=True)
a
```

```python
left_df = pd.DataFrame({'A': [12, 25, 37]}, index=['a', 'b', 'c'])
right_df = pd.DataFrame({'B': [41, 8, 50]}, index=['a', 'b', 'd'])
print(left_df)
print()
print(right_df)
```

```python
pd.merge(left_df, right_df, left_index=True, right_index=True, how='outer')
```

```python
pd.merge(left_df, right_df, left_index=True, right_index=True, how='outer')
```
